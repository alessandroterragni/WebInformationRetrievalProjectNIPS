```python
import pandas as pd
import numpy as np
from pathlib import Path
d = Path().resolve().parent.parent
author_strUniversity = pd.read_csv(str(d) + "/data/generated/author-stringUniversity-new.csv",delimiter=";")
institutions_original = pd.read_csv(str(d) + "/data/generated/institutions_original.csv",encoding = "ISO-8859-1")
institutions = pd.read_csv(str(d) + "/data/generated/institutions.csv",delimiter=';',encoding = "ISO-8859-1")
map_institutions = pd.read_csv(str(d) + "/data/generated/map institutions.csv",delimiter=';')
```

```python
institutions_duplicates = []
for i in range(0,len(author_strUniversity)):
    if(author_strUniversity['University'][i] != " Not in Google Scholar" and author_strUniversity['University'][i] != " "):
        institutions_duplicates.append(author_strUniversity['University'][i])
```

```python
author_university = []
for i in range(0,len(author_strUniversity)):
    if(i%100==0):
        print(i)
    for j in range(0,len(institutions_original)):
        if(author_strUniversity['University'][i] == institutions_original['university_name'][j]):
            author_university.append([author_strUniversity['id'][i],institutions_original['university_id'][j]])
            break
```

```python
author_university_copy = author_university
```

```python
map_institutions_dict = map_institutions.groupby('old_id')['new_id'].apply(list)
```

```python
for i in range(0,len(author_university_copy)):
    if(author_university_copy[i][1] in map_institutions_dict):
        author_university_copy[i][1] = map_institutions_dict[author_university_copy[i][1]][0]
```
